### Анализ системы тэгов

1.1. Система учитывает 10 частей речи: существительное, глагол, прилагательное, местоимение, наречие, адлог, союз, числительное, частица, междометие. Есть еще два тэга: аббревиатура и residual. 
Я бы точно выделила еще, например, числительное-прилагательное ("второй"), потому что оно семантически числительное, но обладает свойствами прилагательного, и ни к одному типу однозначно отнести его нельзя.

1.2. Предполагается, что местоимение попадает в класс местоимений. Но может попасть к прилагательным ("ваш").

1.3. Лемма причастий -- это инфинитив.

1.4. Дал и давал -- к разным; дать и давать.

1.5.  В золотом стандарте такие местоимения помечаются как SPRO, местоименные существительные. Так как у TreeTagger нет такой части речи в тэгах, он помечает их как местоимения.

2.2.1. Все знаки препинания парсятся отдельно, поэтому сокращения типа п. это две строчки. Слова с дефисами воспринимаются как единое целое, грамматические характеристики верны, но лемму программа не выводит. Апостроф заменяется на "&#39". 

2.2.2. У незнакомых слов не определяется лемма, но грамматические характеристики правильные.

2.2.3. Слово "мой" оба раза было распарсено как местоимение, хотя в первом случае это императив. Слово "пока", которое можно интерпретировать как наречие, союз или частицу (прощание), было помечено тэгом R, то есть вообще никак не распаршено. 
То есть, если предлагается какой-то вариант, то он предлагается один -- вероятно, наиболее частотный.

2.2.4. Частеречная омонимия прилагательного и существительного определяется хорошо (столовая). Падежная омонимия у существительных тоже.
